```python
import pandas as pd
import matplotlib.pyplot as plt


# Dataset 1: Study vs Productivity


study_hours = [2,3,4,5,6,1,7,3,4,5]
productivity = [40,50,55,65,70,30,85,48,60,68]

data1 = pd.DataFrame({
    "Study Hours": study_hours,
    "Productivity Score": productivity
})

correlation1 = data1.corr()

print("Correlation Matrix (Study vs Productivity):")
print(correlation1)

plt.figure()
plt.scatter(study_hours, productivity)
plt.xlabel("Study Hours")
plt.ylabel("Productivity Score")
plt.title("Study Hours vs Productivity")
plt.show()



# Dataset 2: Screen Time vs Scores


screen_time = [5,4,6,7,3,8,2,6,7,3]
test_scores = [60,65,55,50,75,45,85,58,52,78]

data2 = pd.DataFrame({
    "Screen Time": screen_time,
    "Test Score": test_scores
})

correlation2 = data2.corr()

print("\nCorrelation Matrix (Screen Time vs Test Score):")
print(correlation2)

plt.figure()
plt.scatter(screen_time, test_scores)
plt.xlabel("Screen Time (Hours)")
plt.ylabel("Test Score (%)")
plt.title("Screen Time vs Test Score")
plt.show()
```
