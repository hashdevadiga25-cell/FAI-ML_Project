```python
import matplotlib.pyplot as plt

# Given values
mean = 55
groups = ['A','B','C']
std_dev = [4,10,15]

# Calculate CV
cv = [(s/mean)*100 for s in std_dev]

print("Coefficient of Variation:")
for g,c in zip(groups,cv):
    print(g,":",round(c,2),"%")

# Plot Graph
plt.figure()
plt.bar(groups,cv)
plt.xlabel("Groups")
plt.ylabel("Coefficient of Variation (%)")
plt.title("Comparison of Coefficient of Variation")
plt.show()
```
